```python
import pandas as pd
import re

train_data = pd.read_csv("data/train_dataset_nan.csv")
```

```python
# Number of characters
train_data['char_count'] = train_data['answers'].apply(len)

# Number of words
train_data['word_count'] = train_data['answers'].apply(lambda x: len(x.split()))
```

```python
java_keywords = [
    'abstract', 'assert', 'boolean', 'break', 'byte', 'case', 'catch', 'char', 'class',
    'const', 'continue', 'default', 'do', 'double', 'else', 'enum', 'extends', 'final',
    'finally', 'float', 'for', 'goto', 'if', 'implements', 'import', 'instanceof', 'int',
    'interface', 'long', 'native', 'new', 'null', 'package', 'private', 'protected',
    'public', 'return', 'short', 'static', 'strictfp', 'super', 'switch', 'synchronized',
    'this', 'throw', 'throws', 'transient', 'try', 'void', 'volatile', 'while'
]

def count_keywords(text):
    return sum(1 for word in re.findall(r'\b\w+\b', text) if word in java_keywords)

train_data['java_keyword_count'] = train_data['answers'].apply(count_keywords)
```

```python
train_data['method_count'] = train_data['answers'].apply(lambda x: len(re.findall(r'\b(public|private|protected)?\s*(static)?\s*\w+\s+\w+\s*\(', x)))
```

```python
train_data['class_count'] = train_data['answers'].apply(lambda x: len(re.findall(r'\bclass\b', x)))
train_data['comment_count'] = train_data['answers'].apply(lambda x: len(re.findall(r'//|/\*|\*/', x)))
```

```python
train_data
```

```python
train_data.to_csv("data/train_data_with_features.csv",index=False)
```
